## import、function定義

```python
import pandas as pd
import numpy as np
import math
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import average_precision_score, precision_score, recall_score
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from matplotlib import pyplot as plt

from keras.models import Sequential 
from keras.layers import Embedding , Dense, Dropout, Flatten
import tensorflow as tf

def f_importances(coef, names):
    imp = coef
    imp,names = zip(*sorted(zip(imp,names)))
    plt.barh(range(len(names)), imp, align='center')
    plt.yticks(range(len(names)), names)
    plt.show()


def RF_training(X , Y):
    X_train = X
    y_train = Y
#     # 處理不平衡資料 
#     ros = RandomOverSampler(random_state = 40)
#     X_train, y_train = ros.fit_resample(X_train, y_train)
    
    # CV 驗證訓練結果是否穩定
    cv = KFold(n_splits=3, random_state=1, shuffle=True)
    model = RandomForestClassifier(n_estimators=50, max_depth=5,random_state=42)
    score = cross_val_score(model,X_train,y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    print ('mean accuracy:',score)
    
    ## 輸出模型
    model.fit(X_train,y_train)    
    return model

def cross_feature(X_catagory):
    ori_columns = X_catagory.columns
    visited = []
    for i in ori_columns:
        for j in ori_columns:
            ## 已經有了 或等於 自身就不做cross
            if j in visited or i == j:
                continue
            name = 'cross_'+str(i)+'_'+str(j)
            X_catagory[name] = X_catagory[i]*X_catagory[j]
        visited.append(i)
```

```python
## 讀取資料
train_x = pd.DataFrame(np.load('train_data.npy'))
train_y = pd.DataFrame(np.load('train_labels.npy'))
test_x = pd.DataFrame(np.load('eval_data.npy'))
test_y = pd.DataFrame(np.load('eval_labels.npy'))
```

## 使用RF初篩特徵

```python
model = RF_training( train_x , train_y )
```

```python
y_test_pred = model.predict(test_x)
print('testing score :' , accuracy_score(test_y, y_test_pred))
confusion_matrix(test_y, y_test_pred)
```

```python
f_importances(model.feature_importances_, test_x.columns)
```

## 特徵工程
- 因test理論上lable為未知，所以讓 test_X 都根據 train_X 來製作特徵

```python
## 移除影響很小的資料
drop_label = train_x.columns[model.feature_importances_ < 0.03]
X_train_refactor = train_x.drop(columns=drop_label, axis = 1)
X_test_refactor = test_x.drop(columns=drop_label, axis = 1)
```

```python
X_train_refactor.columns
```

```python
## 首先增加 24 以mean切分的資料，大於mean為0 小於為1
X_train_refactor['over24'] = (X_train_refactor[24]> train_x.describe()[24][1])*1
X_test_refactor['over24'] = (X_test_refactor[24]> train_x.describe()[24][1])*1
```

```python
model = RF_training( X_train_refactor , train_y )
```

```python
y_test_pred = model.predict(X_test_refactor)
print('testing score :' , accuracy_score(test_y, y_test_pred))
confusion_matrix(test_y, y_test_pred)
```

```python
f_importances(model.feature_importances_, X_test_refactor.columns)
```

```python
X_refactor_with_label = X_train_refactor.copy()
X_refactor_with_label['label'] = train_y
```

```python
import seaborn as sns
# mask with below part
mask_below_24_zero = ((X_refactor_with_label.label == 0) & (X_refactor_with_label.over24 == 0))
mask_below_24_one = ((X_refactor_with_label.label == 1) &  (X_refactor_with_label.over24 == 0))
# plot 
fig, ax = plt.subplots( figsize = (15,9))
ax = sns.distplot(X_refactor_with_label.loc[mask_below_24_zero , 24], kde = False, bins=20, norm_hist = True, label = 'zero')
ax = sns.distplot(X_refactor_with_label.loc[mask_below_24_one , 24], kde = False, bins=20 , norm_hist = True, label = 'one')
ax.legend()
ax.set_title('24 distribution ', fontsize = 20)
```

```python
X_train_refactor['24_below'] = (X_train_refactor[24]< -0.5) *1 
X_train_refactor['24_over'] = (X_train_refactor[24]> 0.5) *1 
X_test_refactor['24_below'] = (X_test_refactor[24]< -0.5) *1 
X_test_refactor['24_over'] = (X_test_refactor[24]> 0.5) *1 
```

```python
# mask with below part
mask_below_49_zero = ((X_refactor_with_label.over24 ==0 ) &(X_refactor_with_label.label == 0))
mask_below_49_one = ((X_refactor_with_label.over24 ==0 ) &(X_refactor_with_label.label == 1))
# plot 
fig, ax = plt.subplots( figsize = (15,9))
ax = sns.distplot(X_refactor_with_label.loc[mask_below_49_zero , 49], kde = False, bins=40, norm_hist = True, label = 'zero')
ax = sns.distplot(X_refactor_with_label.loc[mask_below_49_one , 49], kde = False, bins=40, norm_hist = True, label = 'one')
ax.legend()
ax.set_title('49 vs one and zero when below mean', fontsize = 20)
```

```python
X_train_refactor['49_over_zero'] = (X_train_refactor[49]< 0)* 1 
X_test_refactor['49_over_zero'] = (X_test_refactor[49]< 0)* 1 
```

```python
# mask with below part
mask_below_0_zero = ((X_refactor_with_label.label == 0))
mask_below_0_one = ((X_refactor_with_label.label == 1))
# plot 
fig, ax = plt.subplots( figsize = (15,9))
ax = sns.distplot(X_refactor_with_label.loc[mask_below_0_zero , 0], kde = False, bins=40, norm_hist = True, label = 'zero')
ax = sns.distplot(X_refactor_with_label.loc[mask_below_0_one , 0], kde = False, bins=40, norm_hist = True, label = 'one')
ax.legend()
ax.set_title('0 vs one and zero when below mean', fontsize = 20)
```

```python
X_train_refactor['0_over'] = ((X_train_refactor[0]> 0)) *1 
X_test_refactor['0_over'] = ((X_test_refactor[0]> 0)) *1 
```

```python
# mask with below part
mask_below_18_zero = ((X_refactor_with_label.label == 0))
mask_below_18_one = ((X_refactor_with_label.label == 1))
# plot 
fig, ax = plt.subplots( figsize = (15,9))
ax = sns.distplot(X_refactor_with_label.loc[mask_below_18_zero , 18], kde = False, bins=40, norm_hist = True, label = 'zero')
ax = sns.distplot(X_refactor_with_label.loc[mask_below_18_one , 18], kde = False, bins=40, norm_hist = True, label = 'one')
ax.legend()
ax.set_title('18 vs one and zero when below mean', fontsize = 20)
```

```python
X_train_refactor['18_over'] = (X_train_refactor[18]> 0) *1 
X_test_refactor['18_over'] = (X_test_refactor[18]> 0) *1 
```

```python
## 56 is binary data 
X_train_catagory = X_train_refactor.drop(columns=[24,49,18,0])
X_test_catagory = X_test_refactor.drop(columns=[24,49,18,0])
```

```python
## 對 資料做cross生出交集項
cross_feature(X_train_catagory)
cross_feature(X_test_catagory)
```

```python
X_train_catagory.columns
```

```python
## 對 binary data 取 embedding
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(len(X_train_catagory.columns), 3, input_length=len(X_train_catagory.columns)))
model.add(Flatten())
model.add(Dense(36))
model.add(Dense(3))
model.compile('rmsprop', 'acc')
```

```python
## 將binary資料做embedding，減少訓練的複雜度
train_embedding = model.predict(X_train_catagory.values)
test_embedding = model.predict(X_test_catagory.values)
```

```python
X_train_refactor['embedding_0'] = train_embedding.T[0]
X_train_refactor['embedding_1'] = train_embedding.T[1]
X_train_refactor['embedding_2'] = train_embedding.T[2]
X_test_refactor['embedding_0'] = test_embedding.T[0]
X_test_refactor['embedding_1'] = test_embedding.T[1]
X_test_refactor['embedding_2'] = test_embedding.T[2]
```

```python
model = RF_training( X_train_refactor , train_y )

y_test_pred = model.predict(X_test_refactor)
precision, recall, f1, _ = score(test_y, y_test_pred)
print ( "pre score(0, 1, 2) :" ,precision )    
print ( "rec score(0, 1, 2) :" ,recall )
print ( "f1 score(0, 1, 2)  :" ,f1 )
print('testing accuracy :' , accuracy_score(test_y, y_test_pred))
confusion_matrix(test_y, y_test_pred)

f_importances(model.feature_importances_, X_test_refactor.columns)
```

```python
from xgboost import XGBClassifier

def XGB_training(X , Y):
    X_train = X
    y_train = Y
#     # 處理不平衡資料 
#     ros = RandomOverSampler(random_state = 40)
#     X_train, y_train = ros.fit_resample(X_train, y_train)
    
    # CV 驗證訓練結果是否穩定
    cv = KFold(n_splits=3, random_state=1, shuffle=True)
    model = XGBClassifier(n_estimators=50,random_state=42)
    score = cross_val_score(model,X_train,y_train, scoring='accuracy', cv=cv, n_jobs=-1)
    print ('mean accuracy:',score)
    
    ## 輸出模型
    model.fit(X_train,y_train)    
    return model
```
